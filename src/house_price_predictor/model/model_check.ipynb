```python
import pandas as pd
```

```python

df_raw = pd.read_csv("../../../data/raw/Housing.csv")
print(df_raw.columns)
print(df_raw.head())
```

```python
for col in df_raw.select_dtypes(include='object').columns:
    print(f"{col}: {df_raw[col].nunique()} unique values")
    print(df_raw[col].value_counts())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))
df_raw['price'].hist(bins=30)
plt.title("Distribution of Price")
plt.show()

print("Skewness:", df_raw['price'].skew())
print("Min:", df_raw['price'].min(), "Max:", df_raw['price'].max())
```

```python
# Show the largest and smallest prices
print(df_raw['price'].sort_values().head())
print(df_raw['price'].sort_values(ascending=False).head())
```

```python
import numpy as np

plt.figure(figsize=(8,4))
np.log1p(df_raw['price']).hist(bins=30)
plt.title("Log-Transformed Price Distribution")
plt.show()
```
